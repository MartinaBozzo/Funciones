```python
def factorial(n):
#tu codigo aqu铆
    fact = 1
    for i in range(1, n+1):
       fact*=i
    return fact
```

## 9. Escribe una funci贸n que calcule la desviaci贸n est谩ndar de una lista.
`NOTE: no utilices librer铆as ni ninguna funci贸n ya construida. `

```python
def st_dev(list_sd):
    count = 0
    suma = 0
    
    # Calcular la cantidad de elementos y la suma de los elementos
    for i in list_sd:
        count += 1
        suma += i
    
    # Calcular la media
    media = suma / count
    
    dev = 0
    # Calcular la suma de los cuadrados de las diferencias respecto a la media
    for i in list_sd:
        dev += (i - media) ** 2
    
    # Calcular la varianza
    varianza = dev / count
    
    # Calcular la desviaci贸n est谩ndar como la ra铆z cuadrada de la varianza
    desviacion_estandar = varianza ** 0.5
    
    return desviacion_estandar

```
